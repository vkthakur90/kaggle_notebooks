## Neural Network Example

```python
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
```

### Loading and visualizing dataset

```python
raw_data = pd.read_csv('heart.csv')
```

```python
ax_ = sns.violinplot(x="variable", y="value", data=pd.melt(raw_data))
_ = ax_.set_xticklabels(ax_.get_xticklabels(), rotation = 90)
```

```python
features = raw_data.drop('target', axis = 1)
label = raw_data[['target']]
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(10, activation = 'relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(10, activation = 'relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(10, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
callbacks = model.fit(x = features.values, y = label.values, epochs = 500)
```
